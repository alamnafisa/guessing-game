```python
#Write a code for a Number Guessing Game. The code assigns a random integer number from a given interval to a varible, 
#then asks the user to make a guess to find that number. Check the validity of the selected number:
#Is it an integer?
#Is it an integer that lies in the given range?
#a) Implement a simple version of the game:
#The code assigns a random integer number between 0 and 100 to a variable.
#The code gives the user only one chance to guess that number and the output should be either 'correct' or 'wrong'.

import random
minVal =0
maxVal =100
guess= None
correctnumber= random.randint (minVal,maxVal)
counter=1
numberofTurns= 1 

while counter <= numberofTurns :
     
    try:
        guess= int(input ("Enter an integer value between 0 to 100: "))
        assert minVal <= guess <=maxVal
        
    except ValueError:
        print ("Invalid input! Please enter an integer")
        continue
            
    except AssertionError:
        print ("Invalid range! Enter an integer between 0 and 100")
        continue
    
    if guess == correctnumber:
        print ("Correct!")
        
    else:
        print ("Wrong!") 
            
    counter= counter +1
```
